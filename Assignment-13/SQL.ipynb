**15) Write a sql query to calculate how many books each rate has and have price between 20£ and 30£.**

```python
conn.execute("SELECT COUNT(id),rate FROM books WHERE price between ? AND ? GROUP BY rate",[20,30]).fetchall()
```

**16) Write a sql query to calculate how many books each category has having count > 10.**

```python
conn.execute("SELECT COUNT(id) AS CNT,category_id FROM books GROUP BY category_id HAVING CNT > ?",[10]).fetchall()
```

**17) Write a sql query to get all books with category_name='Music' using subquery.**

```python
import csv
cursor = sqlite3.connect("books.db").cursor()
rows = cursor.execute("SELECT * FROM books")
```

```python
with open('Books.csv','w') as out_csv_file:
  csv_out = csv.writer(out_csv_file)
  # write header                        
  csv_out.writerow([d[0] for d in rows.description])
  # write data                          
  for result in rows:
    csv_out.writerow(result)
```

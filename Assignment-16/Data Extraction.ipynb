note: you can get the converting formula celsius to fahrenheit from google. 

```python
import pandas as pd
from datetime import datetime
import user_agents
import requests
import json
import urllib
import numpy as np
```

```python
log_df = pd.read_csv('datasets/sendy_logistics.csv')
log_df
```

> We need to fill the missing values in Temperature column as we are going to deal with it

```python
log_df['Temperature'].fillna(log_df['Temperature'].mean(),inplace=True)
```

```python
log_df['Temp_Diff'] = round(abs(log_df['Temperature'] - 38),1)
```

```python
log_df['Temp_Diff']
```

```python
def cel_to_fehr(x):
    return (x * 9/5) + 32
```

```python
log_df['Fehr_Temp'] = round(log_df['Temperature'].apply(cel_to_fehr),1)
```

```python
log_df['Fehr_Temp']
```

```python
def in_dec(x):
    return 1 if x == True else 0
```

```python
ufo['IN_December'] = ufo['Time'].dt.month == 12
```

```python
ufo['IN_December'] = ufo['IN_December'].apply(in_dec)
```

> # Task 3:
1) Using visitor-interests.csv dataset, from the first 50 row extract Browser, OS, Device data from 'UserAgent' column and use it to create new columns for Browser, OS, Device.<br>
2) Using visitor-interests.csv dataset, from the first 50 row extract latitude, longitude from 'IP' column.<br>

```python
visitor = pd.read_csv('datasets/visitor-interests.csv')
visitor
```

```python
visitor.info()
```

```python
## functions to extract details
def get_browser(x):
    return user_agents.parse(x).browser.family

def get_os(x):
    return user_agents.parse(x).os.family

def get_device(x):
    return user_agents.parse(x).device.family

def get_lat(x):
    url = f"https://freegeoip.app/json/{x}"
    headers = {
        'accept': "application/json",
        'content-type': "application/json"
        }
    response = requests.request("GET", url, headers=headers)
    respond = json.loads(response.text)
    return respond['latitude']

def get_long(x):
    url = f"https://freegeoip.app/json/{x}"
    headers = {
        'accept': "application/json",
        'content-type': "application/json"
        }
    response = requests.request("GET", url, headers=headers)
    respond = json.loads(response.text)
    return respond['longitude']
```

```python
visitor['Browser'] = visitor['UserAgent'].apply(get_browser)
visitor['OS'] = visitor['UserAgent'].apply(get_os)
visitor['Device'] = visitor['UserAgent'].apply(get_device)
visitor['Latitude'] = visitor['IP'][0:30].apply(get_lat)
visitor['Longitude'] = visitor['IP'][0:30].apply(get_long)
```
